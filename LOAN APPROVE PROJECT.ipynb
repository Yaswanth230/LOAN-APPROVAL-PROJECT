Drop Unnecessary Columns

```python
# Drop ID and ZIP.Code columns
df=df.drop(['ID','ZIP.Code'],axis=1)
```

Define Features and Target

```python
# Define X (independent variables) and y (target)
X=df.drop('Personal.Loan',axis=1)
y=df['Personal.Loan']
```

Train-Test Split

```python
from sklearn.model_selection import train_test_split
# Split data: 70% train, 30% test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42,stratify=y)
# Check shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

üìåImport and Train the Model

```python
from sklearn.tree import DecisionTreeClassifier
# Initialize the model
dt_model=DecisionTreeClassifier(random_state=42)
```

```python
# Train the model on training data
dt_model.fit(X_train,y_train)
```

üìä Make Predictions


```python
# Predict on test data
y_pred=dt_model.predict(X_test)
```

üß™ Evaluate the Model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Accuracy
print("Accuracy:", accuracy_score(y_test,y_pred))

# Classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
# Visualize confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print("Train Accuracy:", dt_model.score(X_train, y_train))
print("Test Accuracy :", dt_model.score(X_test, y_test))
```

üìä Model Evaluation ‚Äì Before Tuning
Train Accuracy: 100%
Test Accuracy: 98.07%
While both scores are high, the training accuracy being exactly 100% is a strong sign of overfitting.
This means the model has likely memorized the training data instead of learning general patterns.

‚ö†Ô∏è Why this is a problem:
Overfitting reduces the model‚Äôs ability to generalize to new, unseen data.
Decision Trees are prone to this when no limits are set on tree depth or leaf size.
‚úÖ Action Plan:
To prevent overfitting, I will use GridSearchCV to tune hyperparameters like:

max_depth
min_samples_leaf
min_samples_split

üì¶ Import and Set Parameter Grid

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Define the parameter grid to search
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize base model
dt = DecisionTreeClassifier(random_state=42)

# Apply GridSearchCV
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5, n_jobs=-1, scoring='accuracy')

# Fit on training data
grid_search.fit(X_train, y_train)
```

```python
GridSearchCV(cv=5, estimator=DecisionTreeClassifier(random_state=42), n_jobs=-1,
             param_grid={'criterion': ['gini', 'entropy'],
                         'max_depth': [3, 5, 10, 15, None],
                         'min_samples_leaf': [1, 2, 4],
                         'min_samples_split': [2, 5, 10]},
             scoring='accuracy')
```

üß™ Use GridSearchCV

```python

# Initialize base model
dt = DecisionTreeClassifier(random_state=42)

# Apply GridSearchCV
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5, n_jobs=-1, scoring='accuracy')

# Fit on training data
grid_search.fit(X_train, y_train)
```

ü•á Best Model and Parameters

```python
print("Best Parameters:", grid_search.best_params_)
```

```python

print("Train Accuracy (Tuned):", grid_search.best_estimator_.score(X_train, y_train))
print("Test Accuracy (Tuned):", grid_search.best_estimator_.score(X_test, y_test))
```

in Accuracy (Tuned): 0.9868571428571429

üîß Model Evaluation ‚Äì After Hyperparameter Tuning
Best Parameters (from GridSearchCV):

criterion: 'gini'
max_depth: 5
min_samples_leaf: 1
min_samples_split: 2
Train Accuracy (Tuned): 98.69%

Test Accuracy (Tuned): 98.47%

‚úÖ Observation:
The training accuracy dropped slightly from 100% to 98.69%, reducing overfitting.
The test accuracy slightly increased from 98.07% to 98.47%, showing improved generalization.
These results confirm that the hyperparameter tuning helped build a more balanced and robust Decision Tree model.

```python

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Best model from GridSearchCV
best_model = grid_search.best_estimator_

plt.figure(figsize=(20,10))
plot_tree(best_model, filled=True, feature_names=X.columns, class_names=["Not Loan", "Loan"], max_depth=3)  # you can adjust max_depth
plt.title("Decision Tree (Tuned Model)")
plt.tight_layout()
plt.show()
```

```python

from sklearn.metrics import classification_report, confusion_matrix

y_pred = best_model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
